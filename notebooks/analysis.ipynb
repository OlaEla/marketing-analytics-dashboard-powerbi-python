```python
import pandas as pd
import os

# Путь к данным
data_path = "../data/processed_data.csv"  # Относительный путь

# Загрузка данных
try:
    df = pd.read_csv(data_path)
    print("Данные успешно загружены!")
    print(f"Количество строк: {len(df)}")
    print("Столбцы:", df.columns.tolist())
except Exception as e:
    print(f"Ошибка загрузки: {e}")
```

```python
# Регионы с максимальными заявками

# Группировка по регионам
top_regions = (
    df.groupby('Region')
    .agg(
        total_requests=('Goal Completions', 'sum'),
        unique_clients=('Client ID', 'nunique')
    )
    .sort_values('total_requests', ascending=False)
    .head(10)
)

print("Топ-10 регионов по заявкам:")
print(top_regions)
```

```python
# Какой средний процент отказов (Bounce)

avg_bounce = df['Bounce Rate'].mean() * 100  # Переводим в проценты

print(f"\nСредний процент отказов: {avg_bounce:.2f}%")

# Дополнительно: динамика отказов по устройствам
bounce_by_device = (
    df.groupby('Device Category')['Bounce Rate']
    .mean()
    .sort_values() * 100
)

print("\nПроцент отказов по устройствам:")
print(bounce_by_device.to_string())
```

```python
# С каких устройств чаще заходят на сайты​

import matplotlib.pyplot as plt
import seaborn as sns

# Расчёт распределения устройств
device_distribution = (
    df['Device Category']
    .value_counts(normalize=True)
    .mul(100)
    .round(1)
)

# Печать распределения в процентах
print("\n📊 Доля посещений по устройствам (%):")
print(device_distribution.to_string())

# Визуализация
plt.figure(figsize=(8, 5))
sns.barplot(
    x=device_distribution.values,
    y=device_distribution.index,
    palette='viridis'
)

plt.title('Доля посещений по категориям устройств (%)', fontsize=14)
plt.xlabel('Доля, %')
plt.ylabel('Категория устройства')
plt.xlim(0, device_distribution.max() + 5)
plt.grid(axis='x', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

Ответьте на вопрос: каких показателей не хватает, чтобы посчитать чистую прибыль?

Чтобы посчитать чистую прибыль, одних данных о продажах, выручке и маржинальности недостаточно. Вот каких показателей не хватает для полного расчёта чистой прибыли:

Необходимые, но отсутствующие показатели:

- Постоянные расходы (фиксированные затраты):
- Аренда офисов/шоурумов
- Зарплаты сотрудников (менеджеров, техников и т.д.)
- Амортизация оборудования
- Административные расходы

Переменные расходы на привлечение клиента:

- Стоимость рекламных кампаний (CPC, CPA, CPM и т.д.)
- Вознаграждения партнёрам, агентствам
- Расходы на маркетинг на одного клиента
- Операционные расходы на продажу авто:
- Стоимость логистики (доставка авто)
- Подготовка авто к продаже (мойка, тех. осмотр, предпродажная подготовка)
- Комиссии дилерам/площадкам

Налоги:

- НДС, налог на прибыль и др.
- Удержания в зависимости от юрисдикции

Прочие затраты:

- Обслуживание IT-инфраструктуры
- Поддержка сайта, CRM, серверов
- Финансовые расходы (например, проценты по кредиту, если авто в лизинг)

Что у нас уже есть:

- Цена продажи авто
- Маржинальность и маржа в рублях
- Выручка в рублях
- Продажи (шт.)
- Конверсии / трафик по источникам

Вывод:
Чтобы посчитать чистую прибыль, нам нужно дополнительно собрать информацию о расходах – как прямых (например, затраты на рекламу), так и косвенных (аренда, ЗП, налоги). Только имея полные данные по затратам, можно получить корректную картину рентабельности бизнеса.

```python
# Сделать прогноз до конца февраля по количеству конверсий на каждый день


# Шаг 1: Преобразуем столбец 'Date' в datetime, если нужно
df['Date'] = pd.to_datetime(df['Date'], errors='coerce')

# Шаг 2: Удалим строки с некорректными датами (если вдруг)
df = df.dropna(subset=['Date'])

# Шаг 3: Нормализуем даты (обнулим время)
df['Date'] = df['Date'].dt.normalize()

# Шаг 4: Группируем по дате и считаем сумму конверсий
daily_conversions = df.groupby('Date')['Конверсия'].sum()

# Шаг 5: Убедимся, что индекс — это datetime
daily_conversions.index = pd.to_datetime(daily_conversions.index)

# Шаг 6: Заполним возможные пропущенные дни (если есть) нулями
daily_conversions = daily_conversions.asfreq('D', fill_value=0)

# Шаг 7: Визуализируем временной ряд
plt.figure(figsize=(14, 5))
plt.plot(daily_conversions, marker='o')
plt.title('Ежедневное количество конверсий')
plt.xlabel('Дата')
plt.ylabel('Конверсии')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
from statsmodels.tsa.holtwinters import SimpleExpSmoothing

# 1. Проверим последнюю дату в данных
last_date = daily_conversions.index.max()
print("Последняя дата в данных:", last_date.date())

# 2. Целевая дата для прогноза
forecast_end_date = pd.to_datetime("2020-02-29")

# 3. Сколько дней нужно спрогнозировать
days_to_forecast = (forecast_end_date - last_date).days
print("Дней до конца февраля:", days_to_forecast)

# 4. Обучение модели экспоненциального сглаживания
model = SimpleExpSmoothing(daily_conversions).fit()

# 5. Прогнозируем
forecast = model.forecast(days_to_forecast)

# 6. Создаём индекс с нужными датами
forecast_index = pd.date_range(start=last_date + pd.Timedelta(days=1), periods=days_to_forecast, freq='D')
forecast = pd.Series(forecast, index=forecast_index)

# 7. Визуализация прогноза вместе с историей
plt.figure(figsize=(14, 5))
plt.plot(daily_conversions[-30:], label='Исторические данные')
plt.plot(forecast, label='Прогноз до 29 февраля', marker='o', linestyle='--')
plt.title('Прогноз количества конверсий до конца февраля 2020')
plt.xlabel('Дата')
plt.ylabel('Конверсии')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# 8. Вывод прогноза
print("\nПрогноз по дням:")
print(forecast.round(0).astype(int))
```

```python
# Для прогнозирования с учетом тренда, используем модель Holt's Linear Trend Model из библиотеки statsmodels. Эта модель будет учитывать тренд и делать прогноз, который будет более динамичным, учитывая изменения в данных.

from statsmodels.tsa.holtwinters import Holt

# Преобразуем столбец 'Date' в индекс и указываем, что дата — это индекс
df['Date'] = pd.to_datetime(df['Date'])
df.set_index('Date', inplace=True)

# Группировка данных по дате для подсчета ежедневных конверсий
daily_conversions = df['Конверсия'].resample('D').sum()

# Обучаем модель Holt с трендом
holt_model = Holt(daily_conversions).fit(smoothing_level=0.8, smoothing_trend=0.2)

# Прогноз на 12 дней вперед (до конца февраля)
forecast_holt = holt_model.forecast(steps=12)

# Выводим прогноз
print(forecast_holt)

# Визуализируем прогноз
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.plot(daily_conversions, label='Исторические данные', color='blue')
plt.plot(forecast_holt, label='Прогноз с учетом тренда', color='red')
plt.xlabel('Дата')
plt.ylabel('Конверсии')
plt.title('Прогноз конверсий с учетом тренда до конца февраля')
plt.legend()
plt.show()
```

Чтобы рассчитать выручку за первый квартал с учетом пропорциональности продаж, выполним следующие шаги:

1. Определим количество дней в первом квартале — это январь, февраль и март. Есть информация о количестве продаж на каждый деньиспользуем это для пропорционального расчета.
2. Определим среднюю цену автомобиля — для этого используем среднюю цену из столбца "Цена в рублях".
3. Определим пропорцию продаж за текущий период (за январь и февраль) и экстраполируем её на количество дней в первом квартале (январь + февраль + март).
4. Рассчитаем выручку — для этого умножаем количество продаж (экстраполированное на первый квартал) на среднюю цену автомобиля.

```python
# 1. Определяем даты для первого квартала (январь, февраль, март)
df['Date'] = pd.to_datetime(df['Date'])  # Убедимся, что даты в правильном формате

# 2. Фильтруем данные для января и февраля
df_q1 = df[df['Date'].dt.month.isin([1, 2])]  # Январь и февраль

# 3. Рассчитываем среднюю цену автомобиля
average_price = df_q1['Цена в рублях'].mean()

# 4. Рассчитываем количество продаж за январь и февраль (можно использовать столбец "Продажа")
total_sales_q1 = df_q1['Продажа'].sum()

# 5. Определяем количество дней в первом квартале
days_in_q1 = 31 + 29 + 31  # Январь (31), февраль (29), март (31) для високосного года

# 6. Определяем количество дней в январе и феврале
days_in_jan_feb = df_q1['Date'].dt.month.isin([1, 2]).sum()

# 7. Пропорционально экстраполируем продажи
sales_per_day = total_sales_q1 / days_in_jan_feb  # Среднее количество продаж в день за январь и февраль
predicted_sales_q1 = sales_per_day * days_in_q1  # Прогнозируем количество продаж за первый квартал

# 8. Рассчитываем выручку за первый квартал
revenue_q1 = predicted_sales_q1 * average_price

print(f"Прогнозируемая выручка за первый квартал: {revenue_q1:,.2f} рублей")
```

```python
import pandas as pd
import numpy as np

# 1. Загружаем объединённый файл
df_sales = pd.read_excel('../data/Premium_Auto_Cleaned.xlsx')

# Приводим к нижнему регистру для надёжного сопоставления
df_sales['Марка'] = df_sales['Марка'].astype(str).str.lower()
df_sales['Модель'] = df_sales['Модель'].astype(str).str.lower()

# 2. Загружаем справочник
sprav = pd.read_excel('../data/df_sprav.xlsx')
sprav['Марка'] = sprav['Марка'].astype(str).str.lower()
sprav['Модель'] = sprav['Модель'].astype(str).str.lower()

# 3. Справочные словари
model_price = dict(zip(zip(sprav['Марка'], sprav['Модель']), sprav['Цена в рублях']))
brand_avg_price = sprav.groupby('Марка')['Цена в рублях'].mean().to_dict()

# 4. Функция для определения цены при продаже
def assign_price(row):
    if row['Продажа'] == 1:
        marka = row['Марка']
        model = row['Модель']
        price = model_price.get((marka, model))
        if pd.notna(price):
            return price
        else:
            return brand_avg_price.get(marka)
    return np.nan

# 5. Применяем
df_sales['Цена при продаже'] = df_sales.apply(assign_price, axis=1)

# 6. Оставляем только нужные столбцы
df_romi = df_sales[['Source conv.', 'Продажа', 'Цена при продаже']].copy()

# 7. Сохраняем упрощённую таблицу
df_romi.to_excel('../data/romi_base.xlsx', index=False)

# 8. Быстрая проверка
print("Финальная таблица для ROMI готова ✅")
print(df_romi.head())
print("Уникальных источников:", df_romi['Source conv.'].nunique())
```

ROMI — это аббревиатура от Return on Marketing Investment, то есть доходность от инвестиций в маркетинг.

Формула классического ROMI:

\[
\textbf{ROMI} = \frac{\text{Доход от продаж} - \text{Маркетинговые расходы}}{\text{Маркетинговые расходы}} \times 100\%
\]

Он показывает, насколько эффективно работают маркетинговые вложения:

ROMI > 0% — реклама окупается,
ROMI < 0% — кампания убыточна,
ROMI = 100% — каждый рубль вложений приносит ещё один рубль прибыли.

Цель ROMI в данном проекте — оценить эффективность маркетинговых каналов, чтобы понять, откуда приходят самые ценные клиенты, и стоит ли усиливать работу с этим источником.

Проблема в нашем случае:
Мы не имеем данных о маркетинговом бюджете, а значит невозможно использовать классическую формулу ROMI.

В таких ситуациях можно использовать альтернативные метрики, которые дают качественную оценку эффективности каналов без затрат:

\[
\textbf{Упрощённый ROMI}_{\text{Источник}} = \frac{\text{Суммарная выручка с источника}}{\text{Количество продаж с источника}}
\]

_Где:_
- **Суммарная выручка с источника** — сумма всех «цен при продаже» клиентов, пришедших из этого источника.
- **Количество продаж с источника** — количество клиентов, совершивших покупку, пришедших из этого источника.

Это позволяет:

- Оценить качество трафика с точки зрения потенциальной выручки
- Сравнить ценность клиентов по разным каналам
- Принять решения о перераспределении бюджета

Преимущества нашего подхода:

| Критерий                  | Описание                                                                 |
|---------------------------|--------------------------------------------------------------------------|
| Нет бюджета               | Классический ROMI невозможен без данных о затратах                       |
| Альтернатива              | Используем выручку и количество продаж как косвенный показатель ROMI     |
| Аналитика каналов         | Сравниваем источники по ценности их трафика                              |
| Поддержка принятия решений| Данные можно использовать для аргументов в пользу изменения активности по каналам |


Мы честно признаём ограниченность данных и обоснованно применяем упрощённую модель ROMI, чтобы:

- Выполнить задачу проекта
- Избежать домыслов о расходах
- Предоставить ценную аналитику по источникам трафика

"Средняя цена продажи" по источнику = альтернативная ROMI-метрика
Чем она выше, тем потенциально качественнее лиды из этого источника

```python
# Загружаем обработанный файл
df_romi = pd.read_excel('../data/romi_base.xlsx')

# Оставляем только строки, где была продажа и известна цена
df_filtered = df_romi[(df_romi['Продажа'] == 1) & (df_romi['Цена при продаже'].notna())]

# Группируем по источнику
romi_simple = df_filtered.groupby('Source conv.').agg({
    'Продажа': 'sum',
    'Цена при продаже': ['sum', 'mean']
})

# Переименовываем столбцы
romi_simple.columns = ['Количество продаж', 'Общая выручка', 'Средняя цена продажи']
romi_simple = romi_simple.sort_values('Средняя цена продажи', ascending=False)

# Сохраняем
romi_simple.to_excel('../data/romi_simple_result_.xlsx')

# Вывод
print("✅ Упрощённый ROMI (доход на 1 продажу по каналу):")
print(romi_simple.head(10).round(0))
```
